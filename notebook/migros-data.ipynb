```python
import requests
import time

def get_place_details(place_id, api_key, fields):
    details_url = "https://maps.googleapis.com/maps/api/place/details/json"
    details_params = {
        'place_id': place_id,
        'fields': ','.join(fields),
        'key': api_key
    }
    response = requests.get(details_url, params=details_params)
    response_json = response.json()

    # Check for and print errors
    if "error_message" in response_json:
        print(f"Error fetching place details: {response_json['error_message']}")
        return None

    return response_json.get("result", {})

def search_places(query, api_key):
    places = []
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    params = {
        'query': query,
        'key': api_key
    }
    response = requests.get(url, params=params).json()
    places.extend(response.get('results', []))

    # Check for and handle pagination if necessary
    next_page_token = response.get('next_page_token')
    while next_page_token:
        time.sleep(2)  # Ensuring the token is valid before using it
        params['pagetoken'] = next_page_token
        response = requests.get(url, params=params).json()
        places.extend(response.get('results', []))
        next_page_token = response.get('next_page_token')

    return places


fields = ["formatted_address", "name", "geometry", "opening_hours", "rating"]

cantons = ["Zürich", "Bern", "Lucerne", "Uri", "Schwyz", "Obwalden", "Nidwalden",
    "Glarus", "Zug", "Fribourg", "Solothurn", "Basel-Stadt", "Basel-Landschaft",
    "Schaffhausen", "Appenzell Ausserrhoden", "Appenzell Innerrhoden", "St. Gallen",
    "Graubünden", "Aargau", "Thurgau", "Ticino", "Vaud", "Valais", "Neuchâtel",
    "Geneva", "Jura"]
all_places = []

for canton in cantons:
    query = f"Coop {canton} Switzerland"
    initial_places = search_places(query, gmaps_key)

    for place in initial_places:
        detailed_info = get_place_details(place['place_id'], gmaps_key, fields)
        all_places.append(detailed_info)
        time.sleep(1)  # Respectful pause to avoid hitting rate limits

print(f"Found {len(all_places)} places with detailed information across cantons.")



```
